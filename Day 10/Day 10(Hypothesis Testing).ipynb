# Testing One tailed T test(As sample is less than 30) 

```python
Sample_SalDataSet = MyData.sample(25)
Mylist = ['Age','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StandardHours','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']
for i in Mylist:
    Pop_mean = SalDataSet[i].mean()
    Pop_std = SalDataSet[i].std()
    Sample_mean=Sample_SalDataSet['MonthlyIncome'].mean()
    Sample_std=Sample_SalDataSet['MonthlyIncome'].std()
    if Sample_SalDataSet['Age'].count() <=30:
        T_Test =(Sample_mean - Pop_mean)/(Sample_std/np.sqrt(25))
        print('T Test Score for  :',i,'is :',T_Test)
    else:
        Z_Test =(Sample_mean - Pop_mean)/(Pop_std/np.sqrt(25))
        print('Z Test Score for  :',i,'is :',Z_Test)
    Alpha = 0.05 
    T_Score = -1.711
    if T_Score < T_Test:
        print("Accept Null hypothesis for: ",i)
    else:
        print("Reject Null hypothesis for: ",i)
print("Completed Hypothesis Testing using One Tailed ")
```
